```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import datasets
```

```python
def step_gradient(data,learning_rate,m,c):
    m_slope, c_slope= 0,0
    N = len(data)
    for i in range(len(data)):
        x = data[i,0]
        y = data[i,1]
        m_slope += (-2/N)*(y-m*x-c)*(x)
        c_slope += (-2/N)*(y-m*x-c)
    new_m = m - learning_rate*m_slope
    new_c = c - learning_rate*c_slope
    return new_m, new_c
```

```python
def gradient_descent(data,learning_rate,iterations):
    m , c = 0, 0
    for i in range(iterations):
        m,c = step_gradient(data,learning_rate,m,c)
        print(i, "Cost: ", cost(data,m,c))
    return m,c
```

```python
def cost(data, m,c):
    tot_cost=0
    N = len(data)
    for i in range(N):
        x = data[i,0]
        y = data[i,1]
        tot_cost += (1/N)*((y-m*x-c)**2)
    return tot_cost
```

```python
def run(learning_rate):
    data = np.array([[0,1], [1,3], [2,2], [3,5], [4,7], [5,8], [6,8], [7,9], [8,10], [9,12]])
    iteration = 100
    m,c = gradient_descent(data,learning_rate,iteration)
    print(m,c)
run(0.001)
```
