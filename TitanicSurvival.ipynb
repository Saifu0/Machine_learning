```python
import pandas as pd
import numpy as np

data = pd.read_csv("training_titanic_x_y_train.csv",delimiter=',')
X_test = pd.read_csv("test_titanic_x_test.csv",delimiter=',')
data.Age.fillna(data.Age.mean(),inplace=True)
data.drop('Name',axis=1,inplace=True)
data.drop('Ticket',axis=1,inplace=True)
data.drop('Cabin',axis=1,inplace=True)
data.drop('Embarked',axis=1,inplace=True)


X_test.Age.fillna(X_test.Age.mean(),inplace=True)
X_test.drop('Name',axis=1,inplace=True)
X_test.drop('Ticket',axis=1,inplace=True)
X_test.drop('Embarked',axis=1,inplace=True)
X_test.drop('Cabin',axis=1,inplace=True)
X_train = data.iloc[:,:-1]
y_train = data.iloc[:,data.shape[1]-1]
# data.Cabin.describe()
# X_train["Cabin"] = X_train.Cabin.fillna('G6')
# X_test["Cabin"] = X_test.Cabin.fillna('G6')
data
```

```python
# X_train["Sex"] = X_train["Sex"].map({"male": 0, "female": 1})
def f(s):
    if s=="male":
        return 1
    else:
        return 0
    
def func(s):
    if s=='S':
        return 0
    elif s=='C':
        return 1
    else:
        return 2
X_train["Sex"] = X_train.Sex.apply(f)

X_test["Sex"] = X_test.Sex.apply(f)
# X_test.Embarked.describe()
X_train
```

```python
y_train.values
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python

y_pred = model.predict(X_test)
print(y_pred)
np.savetxt("y_pred_titanic_survival.csv",y_pred)
```

```python
from sklearn.metrics import classification_report
y_train_pred = model.predict(X_train)
print(classification_report(y_train_pred,y_train))
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_pred,y_train)
```
